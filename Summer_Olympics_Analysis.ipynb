Medals Count

```python
in_London = df[df['City']=='London']
print('Gold :',len(in_London[in_London['Medal']=='Gold']))
print('Silver :',len(in_London[in_London['Medal']=='Silver']))
print('Bronze :',len(in_London[in_London['Medal']=='Bronze']))
```

```python
for Country in df['Country'].unique():
  con = df[df['Country']== Country]
  print(Country,'Gold:',len(con[con['Medal']=='Gold']),'Silver:',len(con[con['Medal']=='Silver']),'Bronze:',len(con[con['Medal']=='Bronze']))
```
